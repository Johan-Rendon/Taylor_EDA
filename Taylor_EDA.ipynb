## Which is the most danceable album

```python
plt.rcParams['figure.figsize'] = [6, 12]
```

```python
albums = df.groupby('album', as_index=False)['danceability']\
        .agg(["mean", "max", "min"])\
        .sort_values(by=['mean'], ignore_index=True)
```

```python
name = albums['album']
maxDan = albums['max']
minDan = albums['min']
meanDan = albums['mean']
```

```python
plt.title('Mean Danceability')
plt.barh(name, meanDan)
```

```python
plt.title('Max vs Min danceability')
plt.barh(name, maxDan)
plt.barh(name, minDan)
```

```python
mostDS = df[['name','danceability']]\
.sort_values('danceability', ignore_index=True, ascending=False)\
.drop_duplicates()\
.head()
```

```python
 danceabilitydf = df\
                .sort_values('danceability', ignore_index=True, ascending=False)\
                .drop_duplicates('name')\
                .drop(columns=['track_number', 'id', 'uri'])
```

```python
danceabilitydf
```

```python
plt.rcParams['figure.figsize'] = [20, 12]
plt.title('Danceability over the time')
plt.grid(True)
danceabilitydf.groupby('release_date')['danceability'].mean().plot(marker='o')
```

```python
plt.figure(figsize=(12,6))
plt.xticks(rotation=90)
sns.lineplot(data=danceabilitydf,
            x= danceabilitydf.sort_values('release_date', ignore_index=True)['album'],
            y= danceabilitydf.sort_values('release_date', ignore_index=True)['danceability'],
            marker='o')
```

```python
sns.heatmap(danceabilitydf.drop(columns=['name', 'album']).corr())
```

```python
plt.rcParams['figure.figsize'] = [10, 15]
plt.xticks(rotation=45)
sns.scatterplot(danceabilitydf.drop(columns=['name', 'album', 'release_date']).corr()['danceability'].drop('danceabilit))
```

```python
labels = ['Not Danceable[0,0.33]', 'Medium[0.33,0.67]', 'Danceable[0.67,1]']
songs_count = []
for i in range(3):
    danceability_lb = i * (1 / 3)
    danceability_ub = (i + 1) * (1 / 3)
    temp_df = df[df['danceability'] > danceability_lb]
    temp_df = temp_df[temp_df['danceability'] <= danceability_ub]
    songs_count.append(temp_df.shape[0]) # Add number of songs to list


plt.figure(figsize=(12,6))
plt.title("Songs danceability distribution")
plt.pie(songs_count, labels=labels, autopct='%1.1f%%', startangle=180)

```
